```python
import numpy as np
import pandas as pd
```

```python
data_frame_cp=pd.read_csv("./country_population.csv")
data_frame_fr=pd.read_csv("./fertility_rate.csv")
data_frame_le=pd.read_csv("./life_expectancy.csv")
cp_copy=data_frame_cp.copy()
fr_copy=data_frame_fr.copy()
le_copy=data_frame_le.copy()
cp_copy
```

```python
# cp_copy.fillna(cp_copy.mean(axis=1)) 
# cp_copy.head(120)
#m = cp_copy.mean(axis=1)

execluded_columns=['Country Name','Country Code','Indicator Name','Indicator Code']
for i,col in enumerate(cp_copy):
    if col in execluded_columns:
        continue
    cp_copy.iloc[:,i].fillna(cp_copy.iloc[:,4:].mean(axis=1),inplace=True)
cp_copy.dropna(inplace=True)

```

```python
fr_copy = fr_copy[~fr_copy['Country Name'].str.contains('Not classified')]

```

```python
execluded_columns=['Country Name','Country Code','Indicator Name','Indicator Code']
for i,col in enumerate(fr_copy):
    if col in execluded_columns:
        continue
    fr_copy.iloc[:,i].fillna(fr_copy.iloc[:,4:].mean(axis=1),inplace=True)
fr_copy.isnull().sum()
```

```python

fr_copy.fillna(0,inplace=True)
fr_copy.isnull().sum()
```

```python
le_copy = le_copy[~le_copy['Country Name'].str.contains('Not classified')]
execluded_columns=['Country Name','Country Code','Indicator Name','Indicator Code']
for i,col in enumerate(le_copy):
    if col in execluded_columns:
        continue
    le_copy.iloc[:,i].fillna(le_copy.iloc[:,4:].mean(axis=1),inplace=True)

le_copy.fillna(0,inplace=True)
le_copy.isnull().sum()
```

```python
new_columns = {str(year): f'{year}_CP' for year in range(1960, 2017)}
cp_copy.rename(columns=new_columns, inplace=True)
new_column_fr = {str(year): f'{year}_FR' for year in range(1960, 2017)}
fr_copy.rename(columns=new_column_fr, inplace=True)
new_column_le = {str(year): f'{year}_LE' for year in range(1960, 2017)}
le_copy.rename(columns=new_column_le, inplace=True)
```

```python

print(cp_copy)
print(le_copy)
print(fr_copy)
cp_copy=cp_copy.drop(['Indicator Name','Indicator Code'],axis=1)
le_copy=le_copy.drop(['Indicator Name','Indicator Code'],axis=1)
fr_copy=fr_copy.drop(['Indicator Name','Indicator Code'],axis=1)
```

```python
print(cp_copy)
print(le_copy)
print(fr_copy)
```

```python
merge_cp_le=pd.merge(cp_copy,le_copy,on=['Country Code','Country Name'])
merged_df = pd.merge(merge_cp_le, fr_copy, on=['Country Code','Country Name'])
merged_df
```

```python
# merged_df.to_csv('merged_df.csv',index=False)
merged_df_copy=merged_df.copy()
```

```python
data_frame_mdc=pd.read_csv("./Metadata_Country.csv")
final_df=pd.merge(merged_df_copy,data_frame_mdc,on='Country Code',how="left")
final_df
```

```python
final_df["Region"] = final_df["Region"].fillna("NA")
final_df.to_csv("final_df.csv",index=False)
```

```python
melted_df = final_df.melt(id_vars=['Country Name', 'Country Code','Region'], var_name='Year_Type', value_name='Value')

# Extract 'Year' and 'Type' from the 'Year_Type' column
melted_df[['Year', 'Type']] = melted_df['Year_Type'].str.extract(r'(\d{4})_(\w{2})')

# Drop the original 'Year_Type' column
melted_df = melted_df.drop(columns='Year_Type')

# Pivot the DataFrame to get 'CP', 'FR', 'LE' as separate columns
pivoted_df = melted_df.pivot_table(index=['Country Name', 'Country Code','Region', 'Year'], columns='Type', values='Value').reset_index()

# Rename the columns if needed
pivoted_df.columns.name = None  # Remove the 'Type' label
pivoted_df = pivoted_df.rename(columns={'CP': 'Country Population', 'FR': 'Fertility Rate', 'LE': 'Life Expectancy'})
```

```python
pivoted_df.to_csv("pivoted_df.csv",index=False)
```

```python
import plotly.express as px

pivoted_df['Country Population'] = pd.to_numeric(pivoted_df['Country Population'], errors='coerce')
filtered_df = pivoted_df[pivoted_df['Region'] != 'NA']


fig=px.scatter(filtered_df, x="Fertility Rate", y="Life Expectancy", animation_frame="Year", animation_group="Country Name",
           size="Country Population", color="Region", hover_name="Country Name",
           log_x=True, size_max=55, range_x=[1,10], range_y=[10,100])
fig.show()
```
